```python
import plotly.express as px
import pandas as pd
```

```python
data = {
    'Olke':['Almaniya','Fransa','Italiya','ABŞ','Turkiye'],
    'Satis':[4500,4664,4312,6754,3215],
    'Menfeet':[1432,1124,987,2453,1145],
    'Region':['Avropa','Avropa','Avropa','Amerika','Avropa']

}

df = pd.DataFrame(data)
```

```python
fig = px.scatter(df, x = 'Satis', y = 'Menfeet',
                 size="Menfeet",
                 color="Region",
                 hover_name="Olke",
                 log_x=False,
                 size_max=60,
                 title="Satis ve Menfeet Analizi (İnteraktiv)")
fig.show()
```
