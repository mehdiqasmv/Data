```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('breast_cancer_by_country.csv')
```

```python
df
```

```python
plt.figure(figsize=(45,30))


sns.barplot(x = "Country",y = "Deaths_2022",data=df,palette='viridis')
```

```python

plt.figure(figsize=(45,30))

fig =px.pie(df,names = 'Country',values='Deaths_2022',title="Top deaths of countries by CORONA VIRUS",hole=0.5)
fig.show()
```

```python
fig = px.scatter(df,
                 x = 'Country',
                 y = 'Deaths_2022',
                 size='Deaths_2022',
                 hover_name='Country',
                 log_x=False,
                 color='Country',
                 size_max=60,
                 title="Top deaths of countries by CORONA VIRUS")
fig.show()
```
